```python
import pandas as pd
df = pd.read_csv('./datasets/game_sales.csv')
```

```python
df.head()
```

```python
# Group by two variables.
pt = df.pivot_table(values="global_sales", index="genre", columns="platform", aggfunc=sum)

pt.head()
```

```python
# Pivot tables are just DataFrames with sorted indexes.

pt.loc["Adventure":"Platform"]
```

```python
pt.loc["Adventure":"Platform", "PC":"PS4"]
```

```python
# axis=index (default) means calculate the statistic across rows.
sumt = pt.sum(axis="index").sort_values(ascending=False)
sumt.head()
```

```python
sumt['PS2':'PS3']
```

```python
sumt['PS2':'PS3'].iloc[0]
```

```python
sumt['PS2':'PS3'].max()
```

```python
sumt['PS2':'PS3'].min()
```

```python
# axis=index means calculate the statistic across columns.

# useful for pivot tables but for dataframes, this doesn't
# makes sense since every column has different data type.

sumt = pt.sum(axis="columns").sort_values(ascending=False)
sumt.head()
```

```python
# Multiple index.
mpt = df.pivot_table(
        values="global_sales", index=["publisher", "platform"],
        columns="year", aggfunc=sum, fill_value=0, margins=True)

mpt = mpt.sort_index() # DONT FORGET TO SORT INDEX
mpt = mpt.loc[:, 2009:]

mpt.head(10)
```

```python
# Slicing by multiple index
mpt = mpt.loc[('Electronic Arts','PC'):('Nintendo','PS4')]
mpt.head(10)
```

```python
# Multiple index and row.
mpt = df.pivot_table(
        values="global_sales", index=["publisher", "platform"],
        columns=["year", "genre"], aggfunc=sum, fill_value=0, margins=True)

mpt = mpt.sort_index() # DONT FORGET TO SORT INDEX
mpt = mpt.loc[:, 2009:]

mpt.head(10)
```

```python
# Slicing by multiple index and columns.
mpt = mpt.loc[
    ('Electronic Arts','PC'):('Nintendo','PS4'),
    (2016, 'Simulation'):(2017, 'Adventure')
]

mpt.head(10)
```
