```python
import pandas as pd
df = pd.read_csv('../datasets/game_sales.csv')
```

```python
df.head()
```

```python
df["Sales"] = df["Global_Sales"] * 1e06 # Convert to million.
df = df.astype({"Sales": int})

df.head()
```

```python
from datetime import date
year = date.today().year

years_passed = year - df["Year"]
df["Sales_Avg_Per_Year"] = df["Sales"] / years_passed

df.sort_values("Sales_Avg_Per_Year", ascending=False).head()
```
