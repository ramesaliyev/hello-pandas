```python
import pandas as pd
df = pd.read_csv('./datasets/game_sales.csv')
```

```python
df.head()
```

```python
# Drop duplicates by the genre column.
df.drop_duplicates(subset="genre").sort_values("genre").head()
```

```python
# Drop duplicates by the genre and platform columns.
df.drop_duplicates(subset=["genre", "platform"]).sort_values("genre").head()
```

```python
# Publishers by most genre.
publisher_genre = df.drop_duplicates(subset=["publisher", "genre"])
publisher_genre[["publisher"]].value_counts(sort=True)
```

```python
# Platforms by most genre.
platform_genre = df.drop_duplicates(subset=["platform", "genre"])
platform_genre[["platform"]].value_counts(sort=True)
```

```python
# Proportions of the dataset by genre.
df[["genre"]].value_counts(sort=True, normalize=True)
```
