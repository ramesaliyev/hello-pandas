```python
import pandas as pd
df = pd.read_csv('./datasets/game_sales.csv')
```

```python
df.head()
```

```python
df.index
```

```python
df.columns
```

```python
# Setting a column as the index
# Values dont need to be unique
df = df.set_index("name")
df.head()
```

```python
# Removing an index
df = df.reset_index()
df.head()
```

```python
# Dropping the index
df_by_names = df.set_index('name')
df_by_names.reset_index(drop=True).head()
```

```python
filtr = ["Super Mario Bros.", "Grand Theft Auto V"]

# Subsetting by column.
ss1 = df[df["name"].isin(filtr)]

# Subsetting by index.
# loc filters on index values.
ss2 = df.set_index('name').loc[filtr]

display(ss1.head())
display(ss2.head())
```

```python
# Multi-level indexes aka Hierarchical Indexes
# Values dont need to be unique and can be multiple.

indexed = df.set_index(['platform', 'genre'])
indexed.head()
```

```python
# Subsetting the outer level index.
indexed.loc[['PC']]
```

```python
# Subsetting the outer level index.
indexed.loc[['PC', 'PS4']]
```

```python
# Subsetting the inner level indexes.
indexed.loc[[('PC', 'Action'), ('PS4','Action')]]
```

```python
indexed.head()
```

```python
# Sorting by index values
indexed.sort_index()
```

```python
# Controlling the sorting.
indexed.sort_index(level=['platform', 'genre'], ascending=[False, False])
```

```python
indexed.sort_index(level=['genre'], ascending=[True])
```
