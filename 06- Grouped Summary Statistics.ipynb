```python
import pandas as pd
df = pd.read_csv('./datasets/game_sales.csv')
```

```python
df.head()
```

```python
# Mean sales for each genre.
df.groupby('genre')['global_sales'].mean()
```

```python
# Mean sales for each platform.
df.groupby('platform')['global_sales'].mean().sort_values(ascending=False)
```

```python
df.groupby('genre')['global_sales'].agg([min, max, sum]).sort_values(by="sum", ascending=False)
```

```python
groupby = ['genre', 'platform']
pick = ['global_sales', 'eu_sales']
agg = [min, max, sum]
sort_by = ['genre', ('global_sales', 'sum')]
sort_asc = [True, False]

df.groupby(groupby)[pick].agg(agg).sort_values(by=sort_by, ascending=sort_asc)
```
